# Latar Belakang
Industri telekomunikasi merupakan salah satu sektor dengan tingkat persaingan yang sangat tinggi. Perusahaan penyedia layanan seperti internet, WiFi, dan telepon harus terus berinovasi, tidak hanya dalam meningkatkan kualitas jaringan, tetapi juga dalam mempertahankan pelanggan yang sudah ada. Salah satu indikator utama yang sering digunakan untuk mengukur keberhasilan strategi retensi pelanggan adalah **churn rate**, yaitu persentase pelanggan yang berhenti menggunakan layanan dalam periode tertentu (Investopedia, 2023).
Churn rate menjadi permasalahan krusial dalam industri telekomunikasi karena pelanggan cenderung bersifat **price-sensitive** dan memiliki banyak alternatif penyedia layanan. Kemudahan berpindah operator, perbedaan harga paket, serta kualitas layanan yang dirasakan pelanggan menjadi faktor utama yang mendorong terjadinya churn. Berdasarkan laporan industri, tingkat churn pada sektor telekomunikasi secara global berada pada kisaran **15% hingga 25% per tahun**, tergantung pada jenis layanan dan segmentasi pelanggan (Statista, 2022).


Tingginya churn rate dapat memberikan dampak negatif yang signifikan terhadap kinerja bisnis perusahaan. Kehilangan pelanggan tidak hanya berarti berkurangnya pendapatan, tetapi juga meningkatkan biaya operasional akibat kebutuhan untuk mengakuisisi pelanggan baru. Beberapa studi menyebutkan bahwa **biaya untuk mendapatkan pelanggan baru dapat mencapai lima hingga sepuluh kali lebih besar dibandingkan mempertahankan pelanggan yang sudah ada** (Harvard Business Review, 2014).
Dalam konteks ini, perusahaan seperti **Telco Technology** yang menyediakan layanan internet, WiFi, dan telepon bagi pelanggan individu maupun bisnis perlu memiliki strategi yang efektif untuk mengurangi tingkat churn. Pendekatan konvensional yang bersifat reaktif, yaitu bertindak setelah pelanggan berhenti berlangganan, sering kali kurang optimal dalam menekan churn rate.


Seiring dengan meningkatnya ketersediaan data pelanggan dan perkembangan teknologi analitik, **machine learning** menjadi pendekatan yang potensial untuk mengatasi permasalahan churn. Dengan memanfaatkan data historis pelanggan, model machine learning dapat digunakan untuk **memprediksi pelanggan yang berisiko churn** berdasarkan pola perilaku, penggunaan layanan, dan karakteristik pelanggan lainnya. Pendekatan prediktif ini memungkinkan perusahaan untuk mengambil tindakan proaktif dalam strategi retensi pelanggan (IBM, 2023).
Oleh karena itu, penelitian ini berfokus pada penerapan metode **machine learning untuk prediksi churn pelanggan** pada perusahaan telekomunikasi, dengan tujuan membantu pengambilan keputusan bisnis berbasis data serta meningkatkan loyalitas pelanggan dalam jangka panjang.

## **Metric Evaluation**


Dalam evaluasi model klasifikasi churn, digunakan **confusion matrix** untuk memahami performa model dalam memprediksi pelanggan yang churn dan tidak churn. Confusion matrix membandingkan hasil prediksi model dengan kondisi aktual pelanggan, yang terdiri dari empat komponen utama:


- **True Negative (TN)**  


  Model memprediksi pelanggan **tidak churn (0)** dan hasil tersebut sesuai dengan kondisi aktual.


- **False Positive (FP)**  


  Model memprediksi pelanggan **churn (1)**, padahal pada kenyataannya pelanggan **tidak churn**.


- **False Negative (FN)**  


  Model memprediksi pelanggan **tidak churn (0)**, padahal pelanggan tersebut **benar-benar churn**.


- **True Positive (TP)**  


  Model memprediksi pelanggan **churn (1)** dan hasil tersebut sesuai dengan kondisi aktual.


### Type of Errors


Dalam konteks churn prediction, terdapat dua jenis kesalahan utama yang perlu diperhatikan:


- **Type I Error (False Positive)**  


  Terjadi ketika model memprediksi pelanggan akan churn, namun sebenarnya pelanggan tersebut tidak churn.  
  Kesalahan ini dapat menyebabkan perusahaan mengambil tindakan retensi yang tidak diperlukan, sehingga berpotensi menimbulkan **pemborosan sumber daya**.


- **Type II Error (False Negative)**  


  Terjadi ketika model memprediksi pelanggan tidak churn, padahal sebenarnya pelanggan tersebut akan churn.  
  Kesalahan ini lebih berisiko karena perusahaan **kehilangan peluang untuk mempertahankan pelanggan**, yang dapat berdampak langsung pada **kehilangan pendapatan**.


### Metric Selection
Dalam permasalahan churn, **False Negative** dianggap lebih kritis dibandingkan False Positive. Oleh karena itu, tujuan utama dari model adalah **meminimalkan False Negative**, sehingga pelanggan yang benar-benar berisiko churn dapat terdeteksi lebih awal.


Untuk mencapai tujuan tersebut, metrik evaluasi utama yang digunakan adalah **F2-Score**, karena metrik ini memberikan bobot yang lebih besar pada **Recall** dibandingkan 


Precision. Dengan demikian, model lebih difokuskan pada kemampuan mendeteksi pelanggan churn sebanyak mungkin.


### F2-Score Formula


$$F2 = \frac{(1 + 2^2) \times (Precision \times Recall)}{(2^2 \times Precision) + Recall}$$


Dimana:


- **Precision**  


\[
Precision = $\frac{TP}{TP + FP}$
\]


- **Recall**  


\[
Recall = $\frac{TP}{TP + FN}$
\]


Keterangan:


- **TP (True Positive)**: Jumlah pelanggan churn yang diprediksi dengan benar  


- **FP (False Positive)**: Jumlah pelanggan tidak churn yang salah diprediksi churn  


- **FN (False Negative)**: Jumlah pelanggan churn yang salah diprediksi tidak churn  


Dengan menggunakan **F2-Score** sebagai metrik utama, model diharapkan mampu mengidentifikasi pelanggan yang berisiko churn secara lebih optimal, sehingga perusahaan dapat mengalokasikan strategi retensi secara lebih efektif dan mengurangi risiko kehilangan pelanggan.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# crosstab kamu
dfContract = pd.crosstab(df["contract"], df["churn"], margins=True)

# --- prepare data ---
ct = dfContract.drop("All").copy()

# pastikan kolom ada
for c in ["No", "Yes"]:
    if c not in ct.columns:
        ct[c] = 0

# urut seperti contoh gambar (Two year di atas)
order = ["Two year", "One year", "Month-to-month"]
ct = ct.reindex([x for x in order if x in ct.index])

no = ct["No"].astype(int)
yes = ct["Yes"].astype(int)
total = (no + yes).astype(int)

no_pct = (no / total * 100).round(1)
yes_pct = (yes / total * 100).round(1)

# diverging values
no_neg = -no.values
yes_pos = yes.values
y = np.arange(len(ct.index))

# --- plot ---
fig, ax = plt.subplots(figsize=(13, 5.5), constrained_layout=True)

bars_no  = ax.barh(y, no_neg, height=0.7, label="Churn = 0 (No)")
bars_yes = ax.barh(y, yes_pos, height=0.7, label="Churn = 1 (Yes)")

ax.axvline(0, color="black", linewidth=1)

# xlim simetris + margin
max_val = max(no.max(), yes.max())
xmax = int(max_val * 1.35)  # margin lebih lega
ax.set_xlim(-xmax, xmax)

# y labels
ax.set_yticks(y)
ax.set_yticklabels(ct.index)
ax.invert_yaxis()  # biar urutan dari atas ke bawah sesuai order

# grid tipis biar readable
ax.xaxis.grid(True, linestyle="--", alpha=0.3)
ax.set_axisbelow(True)

ax.set_title("Contract vs Churn (Two-Sided Barplot)")
ax.set_xlabel("Number of Customers")

# --- label rapi (count + percent) ---
offset = xmax * 0.04  # jarak label dari ujung bar

for i in range(len(ct.index)):
    # kiri (No)
    ax.text(no_neg[i] - offset, y[i],
            f"{no.iloc[i]} ({no_pct.iloc[i]}%)",
            va="center", ha="right", fontsize=10)

    # kanan (Yes)
    ax.text(yes_pos[i] + offset, y[i],
            f"{yes.iloc[i]} ({yes_pct.iloc[i]}%)",
            va="center", ha="left", fontsize=10)

ax.legend(loc="upper right", frameon=True)

plt.show()
```

# INSIGHT

## KESIMPULAN DAN REKOMENDASI BUSINESS

Kesimpulan dari EDA diatas adalah : 

- Contract type merupakan faktor churn paling dominan. 

- Pelanggan dengan kontrak month-to-month memiliki churn rate tertinggi, sedangkan kontrak jangka panjang lebih stabil.

- Metode pembayaran berpengaruh signifikan terhadap churn. 

- Pelanggan dengan electronic check memiliki risiko churn tertinggi, sementara auto-payment menunjukkan churn terendah.

- Paperless billing berkorelasi dengan churn lebih tinggi, menunjukkan bahwa pelanggan yang lebih fleksibel cenderung lebih mudah berpindah.

- Layanan tambahan seperti TechSupport dan OnlineSecurity terbukti menurunkan churn, sementara layanan streaming tidak memberikan dampak signifikan.

- Pelanggan yang churn memiliki monthly charges lebih tinggi di semua jenis kontrak, menunjukkan adanya sensitivitas terhadap harga.

- Competitor merupakan penyebab churn terbesar, terutama karena penawaran yang lebih menarik dan perangkat yang lebih baik.

- Churn juga dipicu oleh faktor internal:

**Attitude**: pengalaman buruk dengan customer support.

**Dissatisfaction**: masalah kualitas produk, jaringan, dan layanan.


Dari **EDA** diatas kita juga bisa melihat **Profil pelanggan dengan risiko churn tertinggi**:

- **Contract: month-to-month**

- **Payment: electronic check**

- **Paperless billing**

- **Fiber optic**

- **Tanpa TechSupport atau OnlineSecurity**

- **Monthly charges tinggi**

## BUSINESS RECOMMENDATION'

1. **Prioritaskan retensi pelanggan month-to-month**

Tawarkan upgrade ke kontrak tahunan atau dua tahunan.

Berikan bundling atau insentif loyalitas untuk pelanggan kontrak fleksibel.

2. **Dorong migrasi ke auto-payment**

Target pelanggan dengan metode electronic check.

Berikan insentif untuk beralih ke pembayaran otomatis.

Tujuannya menurunkan friksi pembayaran dan meningkatkan komitmen pelanggan.

3. **Gunakan service add-on sebagai alat retensi**

Fokus pada penawaran TechSupport dan OnlineSecurity.


Berikan bundling atau free trial untuk pelanggan berisiko tinggi.

4. **Perkuat value proposition untuk melawan kompetitor**

Tawarkan paket bundling dengan device atau benefit tambahan.

Fokus pada peningkatan value, bukan hanya diskon harga.

5. **Tingkatkan kualitas customer support**

Lakukan training berkala untuk staff support.

Perbaiki SOP penanganan komplain.

Fokus pada penyelesaian masalah dalam satu interaksi.

6. **Perbaiki kualitas layanan dan stabilitas jaringan**

Prioritaskan area dengan tingkat komplain tinggi.

Tingkatkan keandalan layanan fiber optic.

Perbaiki kualitas produk dan pengalaman layanan.

7. **Terapkan strategi harga yang lebih terarah**

Hindari diskon massal.

**Gunakan targeted pricing untuk:**

- pelanggan month-to-month

- pelanggan dengan biaya bulanan tinggi

- Tawarkan paket berbasis value.
