## **Metric Evaluation**

Dalam evaluasi model klasifikasi churn, digunakan **confusion matrix** untuk memahami performa model dalam memprediksi pelanggan yang churn dan tidak churn. Confusion matrix membandingkan hasil prediksi model dengan kondisi aktual pelanggan, yang terdiri dari empat komponen utama:

- **True Negative (TN)**  
  Model memprediksi pelanggan **tidak churn (0)** dan hasil tersebut sesuai dengan kondisi aktual.

- **False Positive (FP)**  
  Model memprediksi pelanggan **churn (1)**, padahal pada kenyataannya pelanggan **tidak churn**.

- **False Negative (FN)**  
  Model memprediksi pelanggan **tidak churn (0)**, padahal pelanggan tersebut **benar-benar churn**.

- **True Positive (TP)**  
  Model memprediksi pelanggan **churn (1)** dan hasil tersebut sesuai dengan kondisi aktual.

### Type of Errors

Dalam konteks churn prediction, terdapat dua jenis kesalahan utama yang perlu diperhatikan:

- **Type I Error (False Positive)**  
  Terjadi ketika model memprediksi pelanggan akan churn, namun sebenarnya pelanggan tersebut tidak churn.  
  Kesalahan ini dapat menyebabkan perusahaan mengambil tindakan retensi yang tidak diperlukan, sehingga berpotensi menimbulkan **pemborosan sumber daya**.

- **Type II Error (False Negative)**  
  Terjadi ketika model memprediksi pelanggan tidak churn, padahal sebenarnya pelanggan tersebut akan churn.  
  Kesalahan ini lebih berisiko karena perusahaan **kehilangan peluang untuk mempertahankan pelanggan**, yang dapat berdampak langsung pada **kehilangan pendapatan**.

### Metric Selection

Dalam permasalahan churn, **False Negative** dianggap lebih kritis dibandingkan False Positive. Oleh karena itu, tujuan utama dari model adalah **meminimalkan False Negative**, sehingga pelanggan yang benar-benar berisiko churn dapat terdeteksi lebih awal.

Untuk mencapai tujuan tersebut, metrik evaluasi utama yang digunakan adalah **F2-Score**, karena metrik ini memberikan bobot yang lebih besar pada **Recall** dibandingkan Precision. Dengan demikian, model lebih difokuskan pada kemampuan mendeteksi pelanggan churn sebanyak mungkin.

### F2-Score Formula


$$F2 = \frac{(1 + 2^2) \times (Precision \times Recall)}{(2^2 \times Precision) + Recall}$$


Dimana:

- **Precision**  
\[
Precision = $\frac{TP}{TP + FP}$
\]

- **Recall**  
\[
Recall = $\frac{TP}{TP + FN}$
\]

Keterangan:
- **TP (True Positive)**: Jumlah pelanggan churn yang diprediksi dengan benar  
- **FP (False Positive)**: Jumlah pelanggan tidak churn yang salah diprediksi churn  
- **FN (False Negative)**: Jumlah pelanggan churn yang salah diprediksi tidak churn  

Dengan menggunakan **F2-Score** sebagai metrik utama, model diharapkan mampu mengidentifikasi pelanggan yang berisiko churn secara lebih optimal, sehingga perusahaan dapat mengalokasikan strategi retensi secara lebih efektif dan mengurangi risiko kehilangan pelanggan.

## **Preprocess**

```python
# 1. Pisahkan tipe kolom
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X_train.select_dtypes(include=['object', 'bool']).columns.tolist()

print("Numeric:", numeric_cols)
print("Categorical:", categorical_cols)


# 2. Proses kolom numerik
numeric_process = Pipeline([
    ('fill_na', SimpleImputer(strategy='median')),   # isi missing value
    ('scale', RobustScaler())                        # scaling angka
])


# 3. Proses kolom kategori
categorical_process = Pipeline([
    ('fill_na', SimpleImputer(strategy='most_frequent')),  # isi missing value
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))  # ubah ke angka
])


# 4. Proses kolom ordinal (skor 1â€“5)
ordinal_process = Pipeline([
    ('fill_na', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[[1,2,3,4,5]]))
])


# 5. Gabungkan semua
preprocessor = ColumnTransformer([
    ('num', numeric_process, numeric_cols),
    ('ord', ordinal_process, ['satisfactionscore']),
    ('cat', categorical_process, categorical_cols)
])


# 6. Transform data
X_train_final = preprocessor.fit_transform(X_train)
X_test_final  = preprocessor.transform(X_test)

print("Shape setelah preprocessing:", X_train_final.shape)
```

Untuk memastikan pipeline machine learning yang rapi dan terhindar dari data leakage, preprocessing dilakukan menggunakan ColumnTransformer.

Fitur numerik:

Missing value diisi menggunakan median

Diskalakan dengan RobustScaler untuk mengurangi pengaruh outlier

Fitur kategorikal:

Missing value diisi dengan nilai paling sering muncul

Dikodekan menggunakan One-Hot Encoding

Fitur ordinal (satisfactionscore):

Diperlakukan secara ordinal karena memiliki urutan nilai (1â€“5)

```python


# 1. Prediction

y_pred_model = best_model.predict(X_test)

# tanpa model (rule-based: semua dianggap churn = 1)
y_pred_nomodel = np.ones_like(y_test)



# 2. Confusion Matrix

cm_model = confusion_matrix(y_test, y_pred_model)
cm_nomodel = confusion_matrix(y_test, y_pred_nomodel)



# 3. Hitung kerugian finansial

cost_fp = 10.6
cost_fn = 53

def financial_loss(cm):
    fp = cm[0,1]
    fn = cm[1,0]
    return (fp * cost_fp) + (fn * cost_fn)

loss_model = financial_loss(cm_model)
loss_nomodel = financial_loss(cm_nomodel)

result = pd.DataFrame({
    "Metode": ["Tanpa Model", "Dengan Logistic Regression"],
    "Total Kerugian": [loss_nomodel, loss_model]
})

display(result)



# 4. Plot confusion matrix

fig, ax = plt.subplots(1, 2, figsize=(12,5))

ConfusionMatrixDisplay(cm_model).plot(ax=ax[0])
ax[0].set_title("Dengan Model")

ConfusionMatrixDisplay(cm_nomodel).plot(ax=ax[1])
ax[1].set_title("Tanpa Model")

plt.show()
```
